```python
import json
import random
import time
from datetime import datetime
from uuid import uuid4
```

```python
## Helper Functions
def iso_now():
    return datetime.utcnow().strftime("%Y-%m-%dT%H:%M:%SZ")

def rand_cpu():
    return random.randint(200, 2200)

def rand_mem():
    return random.randint(500, 4500)
```

```python
## Pod Generator
def generate_pod(incident=False):
    status = "Running"
    restarts = random.randint(0, 2)

    if incident:
        status = random.choice(["CrashLoopBackOff", "OOMKilled"])
        restarts += random.randint(2, 5)

    return {
        "status": status,
        "restart_count": restarts,
        "cpu": rand_cpu(),
        "memory": rand_mem(),
    }
```

```python
## Main Log Generator
def generate_log():
    incident = random.random() < 0.3
    pods = {f"pay-pod-{uuid4().hex[:4]}": generate_pod(incident) for _ in range(3)}

    return {
        "cluster": "bayer-prod-eu-west-1",
        "timestamp": iso_now(),
        "incident": incident,
        "services": {
            "payment-service": {
                "pods": pods,
                "error_rate_pct": round(random.uniform(1, 5) if not incident else random.uniform(30, 95), 2)
            }
        }
    }
```

## Continuous Generator (Stop Manually)

```python
OUTPUT = "infra_metrics_stream.jsonl"

while True:
    log = generate_log()
    with open(OUTPUT, "a") as f:
        f.write(json.dumps(log) + "\n")
    print("Log written:", log["timestamp"], "incident:", log["incident"])
    time.sleep(random.uniform(1, 4))
```
